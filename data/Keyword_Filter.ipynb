```python
import pandas as pd
import numpy as np
import os

data_filtered = pd.read_csv('data_100Entry.csv')


Filtered_Keyword = pd.read_csv('/Users/zhongxur/Downloads/599 Project - Sheet1.csv')


df = data_filtered


cancer_key = Filtered_Keyword['cancer']
pregnancy_key = Filtered_Keyword['pregnancy']
alcohol_key = Filtered_Keyword['alcohol']

df['Topic_category_C'] = df['Query'].apply(lambda x:1 if any([k in str(x) for k in cancer_key]) else 0)
df['Topic_category_P'] = df['Query'].apply(lambda x:1 if any([k in str(x) for k in pregnancy_key[:7]]) else 0)
df['Topic_category_A'] = df['Query'].apply(lambda x:1 if any([k in str(x) for k in alcohol_key[:7]]) else 0)


mask = (df['Topic_category_C']== 1) | (df['Topic_category_P']== 1) | (df['Topic_category_A']== 1)

df_category_filtered = df[mask]

df_category_filtered.to_csv('data_topic_filted.csv')
```
